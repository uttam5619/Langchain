```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document
from dotenv import load_dotenv
import json
import os

```

```python
load_dotenv()
secret_key=os.getenv('OPENAI_KEY')

```

```python
# importing product.json and storing the data into the product_list
product_list=[]
with open("../data/products.json", "r") as file:
    data = json.load(file)
    product_list=data["product_list"]

print(product_list)
```

```python
# creating Document objects doc for each individual product and storing them into the docs list.
docs=[]
for product in product_list:
    doc = Document(
        page_content=product["description"],
        metadata={
            "id": product["id"],
            "name": product["name"],
            "category": product["category"],
            "brand": product["brand"],
            "features": product["features"],
            "price": product["price"],
            "rating": product["rating"]
        }
    )
    docs.append(doc)
print(docs)
```

```python
# configuring the vector store
embeddings = OpenAIEmbeddings(
    openai_api_key=secret_key
)

vectorStore = Chroma(
    embedding_function= embeddings,
    persist_directory='chroma',
    collection_name='products'
)
```

```python
# adding the documents in vector db
vectorStore.add_documents(docs)
```

```python
# view documents
vectorStore.get(include=['embeddings','metadatas','documents'])
```

```python
# search the vectorDb
result= vectorStore.similarity_search(
    query='apple products',
    k=2 # top query 
)
print(result)
```

```python
# search with similarity score
result_with_score=vectorStore.similarity_search_with_score(
    query='televison',
    k=3
)
print(result_with_score)
```

```python
# metadata filtering
vectorStore.similarity_search_with_score(
    query='high resolution display',
    filter={'brand':'Samsung'}
)
```
